# Quick plotting
A quick notebook to allow rapid plotting of new points for exploration

```python
import numpy as np
import lrlx_plot as lrlx
import matplotlib.pyplot as plt
```

```python
NEW_SRC = {'NEW_SRC_NAME': 'New source',
           'NEW_SRC_LX': [5e33,1e34],
           'NEW_SRC_LX_ler': [4e33,8e33],
           'NEW_SRC_LX_uer': [5e33,9e33],
           'NEW_SRC_LR': [1e29,1e30],
           'NEW_SRC_LR_ler': [9e28,7e29],
           'NEW_SRC_LR_uer': [8e28,6e29]}


def new_plot(NEW_SRC_NAME, 
             NEW_SRC_LX, NEW_SRC_LX_ler, NEW_SRC_LX_uer,
             NEW_SRC_LR, NEW_SRC_LR_ler, NEW_SRC_LR_uer):
    DATA = lrlx.data_reader()
    FIG = plt.figure(figsize=(8,6))
    FIG = lrlx.plotter(DATA, 
                       ['BH', 'candidateBH', 
                        'NS', 'candidateNS', 
                        'AMXP', 'tMSP', 'WD'],
                       FIG, cor_lines=True)
    AX = FIG.axes[0]
    newsrc = AX.errorbar(NEW_SRC_LX, NEW_SRC_LR,
                         xerr=[NEW_SRC_LX_ler,NEW_SRC_LX_uer],
                         yerr=[NEW_SRC_LR_ler,NEW_SRC_LR_uer],
                         fmt='cp', ms=8, ecolor='k', elinewidth=0.8,
                         label=NEW_SRC_NAME)
    AX.legend()
    return FIG

FIG = new_plot(**NEW_SRC)
#FIG.savefig('newsrc_lrlx_plot.jpg',dpi=300, bbox_inches='tight')
```
